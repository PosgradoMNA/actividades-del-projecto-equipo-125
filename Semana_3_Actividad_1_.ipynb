##Parte 1: Fundamentos de bases de datos

###Fundamentos de bases de datos y para ciencia de datos.  

###Limpieza de datos
La limpieza de datos es un proceso que implica la corrección o eliminación de registros de datos erróneos de un conjunto o base de datos. Este proceso de limpieza de datos permite identificar datos: incompletos, incorrectos, inexactos, no pertinentes, entre otros. Una vez identificados, se pueden substituir, modificar o eliminar. En resumen, se refiere una actividad que nos permite preparar un conjunto de datos para posteriormente ejecutar modelos, donde previemante se eliminan, se corrigen datos erróneos que pudieron incluirse en la base de datos de forma accidental, se completan o se agregan. Todo ello empleando alguna técnica de manejo y gestión de datos.

###Tipos de datos faltantes
a) Datos nulos son aquellos que están vacíos.
b) Datos con errores (por ejemplo, que rebasan el intervalo establecido o que hubo error en su captura, o que no cumplen con el formato establecido.

###¿Cómo manejar los datos faltantes?
Para manejar los datos faltantes se pueden completar los registros, eliminarlos, interpolar los datos, entre otros. 

###Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

###Data Warehouse
Los almacenes de datos tienen una gran importancia en las empresas de la actualidad. Es ahí donde se juntan los datos de distintas fuentes (pueden ser datos relacionales, historicos, o datos que se actualizan en tiempo real) y se transforman para que finalmente puedan ser analizados utilizando SQL o herramientas visuales como Tableau. "Un almacén de datos puede considerarse la **única fuente de verdad** de una organización” (Oracle, 2022). William Inmon describió 4 características únicas de estos: 
orientados al sujeto, integrados, no volátiles y variables con el 
tiempo.

###De ETL a ELT 
La forma convencional de hacer en almacenamiento de datos era extrayendo los datos, transformándolos, generalmente por medio de una normalización que ayudara a tenerlos relacionados y estructurados y finalmente cargándolos. Sin embargo, conforme más 
información se recaba, las transformaciónes se vuelven un problema. POr ello se cambia del ETL al ELT, cambiando el orden de los procesos. Primero se cargan los datos y después se transforman. Eso se logra gracias al poder que hoy en día tienen los almacenes de datos. "Eso puede terminar siendo más simple y escalable, y aprovechando el poder de los grandes clústeres de computación que están disponibles en la actualidad" (Kane, 2017).

###Cloud Data Warehouse
Aunque los almacenes de datos tuvieron su origen en servidores locales, actualmente los almacenes en la nube tiene muchas ventajas como "flexibilidad, escalabilidad, agilidad, seguridad y costos reducidos"(Oracle, 2022).

```python
#Guardamos los índices de estos datos faltantes en la columna "X23"
nulos=(246,247)
```

```python
#Hacemos una lista de las columnas que tienen datos faltantes y los contamos para saber cuántos son y ver si podría ser posible eliminar alguna columna
missings=[]
for col in ndf:
  if (ndf[col].isna().sum() != 0):
    print(col, ndf[col].isna().sum()) #Imprime cuántos datos faltantes hay por columnas.
    missings.append(col)
```

###Reemplazando datos nulos por los valores de la mediana
Como son pocos datos faltantes, se puede implementar un método más sencillo que emplear una regresión lineal que consuma más recursos para encontrar cada dato.

```python
#Si no sabemos si hay datos atípicos, se puede emplear mejor la mediana que el promedio.
#Ya que el promedio es muy sensible a datos atípicos.
#Como en realidad son pocos los datos faltantes de cada columna, optamos por 
#reemplazarlos con la mediana. Por ello creamos el siguiente DataFrame
#Se crea una lista donde se tenga la columna, promedio, mediana, el mínimo, el máximo y la desviación estándar.
Col=[]
Mean=[]
Median=[]
Min=[]
Max=[]
Std=[]
#Por cada una de las columnas donde exitan datos faltantes, agrega a la lista de columna el promedio, mediana, mínimo, máximo y desviación estándar. 
for col in missings:
  Col.append(col)
  Mean.append(np.mean(ndf[col]))
  Median.append(ndf[col].median(skipna=True))
  Min.append(np.min(ndf[col]))
  Max.append(np.max(ndf[col]))
  Std.append(ndf[col].std())
#Todas esas listas las pone como olumnas en un DataFrame para una mejor visualización.
mis_data=pd.DataFrame(list(zip(Col,Mean,Median,Min,Max,Std)), columns=["Col","Mean","Median","Min","Max","Std"])
mis_data["Range"]=mis_data["Max"]-mis_data["Min"]
mis_data
```

```python
#Los índices donde la columna 12 tiene datos nulos
#De la 12 a las 17 es el monto de la deuda en dólares (de septiembre hasta abril).
#Justo esos renglones coincide con que todos los datos son nulos.
x=ndf["X12"].isnull()
x=ndf[x]
x
x.loc[:,"X12":"X17"] #Imprime el DatFrame con esos índices.
#x6=pd.DataFrame(list(zip(x,x1,x2,x3,x4,x5)), columns=["Sept","Ago","Jul","Jun","May","Abr"])
#ndf[x6]
```

```python
#No se utiliza el promedio ya que es sensible a datos atípicos (o muy grandes)
#Se opta por reemplazar los datos nulos por los valores de la mediana con la lista de datos faltantes que hicimos anteriormente.
for miss_data in missings:
  ndf[miss_data].fillna(value = ndf[miss_data].median(), inplace = True)
ndf.isna().values.any()
#Ya no hay datos nulos.
```
