###Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

###Data Warehouse
Los almacenes de datos tienen una gran importancia en las empresas de la actualidad. Es ahí donde se juntan los datos de distintas fuentes (pueden ser datos relacionales, historicos, o datos que se actualizan en tiempo real) y se transforman para que finalmente puedan ser analizados utilizando SQL o herramientas visuales como Tableau. "Un almacén de datos puede considerarse la **única fuente de verdad** de una organización” (Oracle, 2022). William Inmon describió 4 características únicas de estos: 
orientados al sujeto, integrados, no volátiles y variables con el 
tiempo.

###De ETL a ELT 
La forma convencional de hacer en almacenamiento de datos era extrayendo los datos, transformándolos, generalmente por medio de una normalización que ayudara a tenerlos relacionados y estructurados y finalmente cargándolos. Sin embargo, conforme más 
información se recaba, las transformaciónes se vuelven un problema. POr ello se cambia del ETL al ELT, cambiando el orden de los procesos. Primero se cargan los datos y después se transforman. Eso se logra gracias al poder que hoy en día tienen los almacenes de datos. "Eso puede terminar siendo más simple y escalable, y aprovechando el poder de los grandes clústeres de computación que están disponibles en la actualidad" (Kane, 2017).

###Cloud Data Warehouse
Aunque los almacenes de datos tuvieron su origen en servidores locales, actualmente los almacenes en la nube tiene muchas ventajas como "flexibilidad, escalabilidad, agilidad, seguridad y costos reducidos"(Oracle, 2022).

##Parte 2: Selección y limpieza de los Datos en Python

```python
#importar librerías
import pandas as pd
import numpy as np
```

```python
#importar data set
input = "https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
```

```python
#convertir a DataFrame y ver las primeras filas
df = pd.read_csv(input)
df.head()
```

###Buscando datos nulos

```python
#Vemos si hay datos nulos en el df
df.isnull().values.any()
```

```python
#Vemos qué columnas tienen datos nulos y cuántos
sd=pd.DataFrame(df.isna().sum())
sd
```

```python
#Porcentaje de datos faltantes
Datos_faltantes=np.sum(sd[0])
porcentaje = np.sum(sd[0])/df.shape[0]*100
Dos_porcentaje = "{:.2f}".format(porcentaje)
print("Datos faltantes:",Datos_faltantes,"\n" ,"Esto es un: ",Dos_porcentaje,"%")
```

```python
#Vemos la cantidad de filas y columnas del df
df.shape
```

```python
#Creamos una copia del df
ndf=df.copy()
```

###Eliminando algunos datos

```python
#eliminamos aquellas filas que tengan más de un 52% de datos nulos, es decir que tengan solamente 13 datos no nulos en toda la fila
ndf.dropna(thresh=(25-12), inplace = True)
ndf.shape
```

```python
#Vemos la cantidad de datos nulos por columna. 
for col in ndf:
  if (ndf[col].isna().sum() != 0):
    print(col, ndf[col].isna().sum())
```

```python
#Buscamos el índice que dato nulo de la columna objetivo Y
x=ndf['Y'].isnull()
x
ndf[x]
```

```python
#Eliminamos la fila donde el dato target "Y" es nulo
ndf = ndf.drop(6232)
```

```python
#Vemos un par de valores de la columna "X23" para después comparar nuestros resultados
x=ndf["X23"].isnull()
x
ndf[x]
```

```python
#Guardamos los índices de estos datos faltantes en la columna "X23"
nulos=(246,247)
#182,219,239,
```

```python
#Hacemos una lista de las columnas que tienen datos faltantes y los contmos para
# saber cuántos son y ver si podría ser posible eliminar alguna columna
missings=[]
for col in ndf:
  if (ndf[col].isna().sum() != 0):
    print(col, ndf[col].isna().sum())
    missings.append(col)
```

###Reemplazando datos nulos por los valores de la mediana

```python
#Como en realidad son pocos los datos faltantes de cada columna, optamos por 
#reemplazarlos con la media o mediana. Por ello creamos el siguiente DataFrame
Col=[]
Mean=[]
Median=[]
Min=[]
Max=[]
Std=[]

for col in missings:
  Col.append(col)
  Mean.append(np.mean(ndf[col]))
  Median.append(ndf[col].median(skipna=True))
  Min.append(np.min(ndf[col]))
  Max.append(np.max(ndf[col]))
  Std.append(ndf[col].std())

mis_data=pd.DataFrame(list(zip(Col,Mean,Median,Min,Max,Std)), columns=["Col","Mean","Median","Min","Max","Std"])
mis_data["Range"]=mis_data["Max"]-mis_data["Min"]
mis_data
```

```python
#Ahora que vemos que hay promedios que se desplazan por datos muy grandes en
#algunas columnas, optamos por poner reemplazar los datos nulos por los valores 
#de la mediana con la lista de datos faltantes que hicimos anteriormente.
for miss_data in missings:
  ndf[miss_data].fillna(value = ndf[miss_data].median(), inplace = True)
ndf.isna().values.any()
```

```python
#checamos los nuevos valores utilizando los índices de valores nulos en la 
#columna 23
for n in nulos:
  print(ndf.iloc[n,23])
```
